# Initial Data Catalog EDA

This notebook starts the v0.4.0 exploration phase.

It reviews currently available project assets before any real ML training starts.


## Scope

This notebook focuses on:

- notebook data catalog readiness
- sample Malaysia locations
- weather summary availability
- geospatial readiness reports
- dataset/training readiness blockers

It does not train a model.


```python
import json
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().resolve()
PROJECT_ROOT
```

```python
from floodrisk.notebooks.catalog import build_notebook_data_catalog

catalog = build_notebook_data_catalog(PROJECT_ROOT)
catalog.as_dict()
```

```python
catalog_rows = []

for entry in catalog.entries:
    catalog_rows.append(
        {
            "asset_id": entry.asset.asset_id,
            "label": entry.asset.label,
            "category": entry.asset.category,
            "path": entry.asset.relative_path,
            "required_for_eda": entry.asset.required_for_eda,
            "status": entry.status,
            "rows": entry.row_count,
            "size_bytes": entry.file_size_bytes,
        }
    )

catalog_df = pd.DataFrame(catalog_rows)
catalog_df
```
